# OG Web.site Unit 3 - Statistical Analysis Practical Evidence

This notebook uses the local `titanic.csv` file from the active Unit 3 folder. It supports the branded evidence workflow by showing the code, output and short annotations in one place.


```python
# Import Pandas and locate the Titanic CSV file.
# The active Unit 3 folder stores titanic.csv at the top level.
from pathlib import Path
import pandas as pd

csv_path = Path('titanic.csv')
if not csv_path.exists():
    csv_path = Path('data') / 'titanic.csv'

titanic_df = pd.read_csv(csv_path)
print('Loaded file:', csv_path)
print('Rows and columns:', titanic_df.shape)
```

```python
# Inspect the DataFrame structure before analysing it.
# This evidence proves that the CSV has loaded and shows the available columns.
print('First five rows:')
display(titanic_df.head())

print('Column data types:')
display(titanic_df.dtypes)
```

```python
# Produce summary statistics for numeric columns.
# describe() gives count, mean, standard deviation, min, quartiles and max.
display(titanic_df.describe())
```

```python
# Calculate individual statistical measures for clear evidence screenshots.
numeric_columns = ['Age', 'Fare', 'SibSp', 'Parch', 'Survived']

print('Mean values:')
display(titanic_df[numeric_columns].mean(numeric_only=True))

print('Median values:')
display(titanic_df[numeric_columns].median(numeric_only=True))

print('Mode values:')
display(titanic_df[numeric_columns].mode().head(1))

print('Standard deviation values:')
display(titanic_df[numeric_columns].std(numeric_only=True))
```

```python
# Grouped analysis compares categories instead of only showing whole-table averages.
# This makes the evidence stronger because it answers practical questions from the data.
print('Survival rate by sex:')
display(titanic_df.groupby('Sex')['Survived'].mean())

print('Survival rate by passenger class:')
display(titanic_df.groupby('Pclass')['Survived'].mean())
```

```python
# Check missing values so the evidence includes data quality inspection.
print('Missing values by column:')
display(titanic_df.isna().sum())

# Practical filter: passengers under 5 who survived.
young_survivors = titanic_df[(titanic_df['Age'] < 5.0) & (titanic_df['Survived'] == 1)]
display(young_survivors.head())
```
