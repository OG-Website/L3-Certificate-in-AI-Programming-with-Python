# Step 1: Create a DataFrame


```python
import pandas as pd

# Step 1: Create a DataFrame and load the CSV file into it
titanic_df = pd.read_csv("data/titanic.csv")
```

# Step 2: Access a DataFrame

```python
# Step 2: Access and explore the dataset

# Display the first 5 rows of the DataFrame
print("First 5 rows of the dataset:\n")
print(titanic_df.head())

# Check data types of columns
print("\nData types for each column:\n")
print(titanic_df.dtypes)

# Summarise the dataset
print("\nSummary statistics:")
print(titanic_df.describe())
```

# Step 4: Deleting elements within a DataFrame
You can delete elements within DataFrames in a number of ways, here are a few: -
* Drop Rows or Columns: You can use the drop() method to remove rows or columns from a DataFrame. This method creates a new DataFrame with the specified rows or columns removed.
* Filtering: You can filter rows based on specific conditions using boolean indexing.

*Note: If "Survived" is 1, it typically means the passenger survived. If "Survived" is 0, it typically means the passenger did not survive, indicating that they died.*

## Example 1 - drop columns

```python
# Let's create a new DataFrame with the specified column removed
drop_df = titanic_df.drop(columns=['Cabin'])

# Display the new DataFrame to check "cabin" column has been removed
print("Has the column 'Cabin' been removed?:\n")
print(drop_df.head())
```

## Example 2 - filtering

```python
# Let's create a new DataFrame with the specified rows or columns removed.
filtered_df = titanic_df[(titanic_df['Age'] < 5.0) & (titanic_df['Survived'] == 1)]

# Display the new DataFrame to check the filtering has been applied
print("Has the filer been applied?:\n")
print(filtered_df)
```
