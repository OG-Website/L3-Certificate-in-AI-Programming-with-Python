# Pandas Series
## How to create, access, and delete elements within Series.

# Step 1: Create a Series


```python
# Create a Series from a list
import pandas as pd
my_list = [10, 20, 30, 40, 50]
series_from_list = pd.Series(my_list)
print(series_from_list)
```

```python
# Create a Series from a NumPy array
import numpy as np
my_array = np.array([1, 2, 3, 4, 5])
series_from_array = pd.Series(my_array)
print(series_from_array)
```

```python
# Create a Series from titanic dataset
import pandas as pd

# Load the Titanic dataset into a DataFrame
titanic_df = pd.read_csv("data/titanic.csv")

# Extract a specific column as a Series
# For example, let's extract the 'Age' column as a Series
age_series = titanic_df['Age']

# Display the Series
print(age_series)
```

# Step 2: Access a Series

```python
# Accessing Elements by Index Label
import pandas as pd

# Create a Series
data = {'A': 10, 'B': 20, 'C': 30, 'D': 40}
series = pd.Series(data)

# Access an element by index label
print(series['A']) 
```

```python
#Accessing Elements by Position using integer-based positional indexing i.e. starts with zero
print(series[0])  
```

```python
# Slicing by index labels
print(series['A':'B'])  

# Slicing by position
print(series[2:4]) 
```

### Step 3: Deleting elements within a Series
You can delete elements within Series in a number of ways, here are a few: -
* Drop: You can use the drop() method to delete elements by specifying the index label(s) of the elements you want to remove.
* Slicing: You can use slicing to select on the elements you want to keep. 

## Example 1 - drop

```python
import pandas as pd

# Create a Series
data = {'A': 10, 'B': 20, 'C': 30, 'D': 40}
series = pd.Series(data)

# Delete elements with index labels 'B' and 'C'
series = series.drop(['B', 'C'])

print(series)
```

## Example 2 - slicing

```python
import pandas as pd

# Create a Series
data = {'A': 10, 'B': 20, 'C': 30, 'D': 40}
series = pd.Series(data)

# Delete elements with index labels 'B' and 'C' and 'D' using slicing
series = series[:'A']

print(series)
```
