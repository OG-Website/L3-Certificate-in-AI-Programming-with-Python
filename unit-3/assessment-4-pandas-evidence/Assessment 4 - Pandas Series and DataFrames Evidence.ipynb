# Unit 3 Assessment 4 - Pandas Series and DataFrames

This notebook is practical evidence for creating, accessing and deleting elements in Pandas Series and DataFrames. The comments explain each step so the screenshots can be used as assessment evidence.

```python
import pandas as pd

print("Pandas version:", pd.__version__)
```

## Series - create elements

A Pandas Series is a one-dimensional labelled data structure. Here I create a Series from score values and give each value a label using the index.

```python
# Create a Pandas Series from a Python list.
# The index labels identify each element by name instead of only by position.
og_series = pd.Series([82, 76, 91], index=["Planning", "Build", "Testing"], name="OG Website Scores")

print(og_series)
print("Series index:", list(og_series.index))
```

## Series - access elements

Series values can be accessed by label or by position. This proves that I can retrieve individual elements from the Series.

```python
# Access by label using .loc[].
planning_score = og_series.loc["Planning"]

# Access by position using .iloc[].
first_score = og_series.iloc[0]

print("Planning score by label:", planning_score)
print("First score by position:", first_score)
```

## Series - delete elements

The drop() method removes a labelled element from a Series. I store the result in a new Series so the original evidence remains clear.

```python
# Delete the Testing element by label.
series_after_delete = og_series.drop("Testing")

print("Original Series:")
print(og_series)
print("\nSeries after deleting Testing:")
print(series_after_delete)
```

## DataFrame - create elements

A DataFrame is a two-dimensional labelled data structure. It works like a table with rows and columns. Here I create a small OG Web.site assessment evidence table.

```python
# Create a Pandas DataFrame from a dictionary.
# Each key becomes a column, and each list supplies the column values.
og_dataframe = pd.DataFrame({
    "Task": ["Notebook", "Screenshots", "GitHub Upload"],
    "Library": ["Pandas", "Pandas", "Pandas"],
    "Status": ["Created", "Captured", "Uploaded"],
    "Score": [82, 76, 91]
})

print(og_dataframe)
```

## DataFrame - access elements

DataFrame values can be accessed by column name, row position, or row and column together. I use column access, .loc[] and .iloc[] to show different ways of retrieving data.

```python
# Access one column by name.
status_column = og_dataframe["Status"]

# Access a row by position.
second_row = og_dataframe.iloc[1]

# Access one cell by row and column label.
first_task_status = og_dataframe.loc[0, "Status"]

print("Status column:")
print(status_column)
print("\nSecond row:")
print(second_row)
print("\nFirst task status:", first_task_status)
```

## DataFrame - delete elements

The drop() method can delete columns or rows. The axis value controls whether Pandas removes a column or a row.

```python
# Delete the Score column by setting axis=1.
dataframe_without_score = og_dataframe.drop("Score", axis=1)

# Delete the row with index 1 by setting axis=0.
dataframe_without_row = og_dataframe.drop(1, axis=0)

print("DataFrame after deleting Score column:")
print(dataframe_without_score)
print("\nDataFrame after deleting row index 1:")
print(dataframe_without_row)
```
