# Unit 4 Assessment 1 - Vector Evidence

This notebook demonstrates vectors, transpose, magnitude, direction, vector addition, scalar multiplication, dot product, and how vector operations support Python programming.

```python
import numpy as np

print("NumPy version:", np.__version__)
```

## Create vectors and check their structure

A vector is an ordered list of numerical components. Here I create two NumPy vectors and check the shape of the first vector.

```python
# Create two vectors as NumPy arrays.
# Each value is a component of the vector.
vector_a = np.array([4, 3])
vector_b = np.array([1, 2])

print("Vector A:", vector_a)
print("Vector B:", vector_b)
print("Vector A shape:", vector_a.shape)
```

## Transpose a vector

Transpose changes the orientation of a vector. A row vector can become a column vector while keeping the same values.

```python
# Create a two-dimensional row vector.
row_vector = np.array([[4, 3]])

# Transpose changes the row vector into a column vector.
column_vector = row_vector.T

print("Row vector:")
print(row_vector)
print("Column vector after transpose:")
print(column_vector)
print("Row shape:", row_vector.shape)
print("Column shape:", column_vector.shape)
```

## Calculate magnitude and direction

Magnitude is the length of the vector. Direction can be shown as an angle from the positive x-axis.

```python
# np.linalg.norm() calculates the vector magnitude.
magnitude = np.linalg.norm(vector_a)

# arctan2(y, x) calculates the angle in radians.
angle_radians = np.arctan2(vector_a[1], vector_a[0])
angle_degrees = np.degrees(angle_radians)

print("Magnitude of Vector A:", magnitude)
print("Direction of Vector A in degrees:", round(angle_degrees, 2))
```

## Vector addition and multiplication

Vector addition adds matching components. Scalar multiplication scales the vector. Dot product multiplies matching components and adds the results.

```python
# Add matching components: [4, 3] + [1, 2] = [5, 5].
vector_sum = vector_a + vector_b

# Multiply every component by 2.
scaled_vector = 2 * vector_a

# Dot product: (4 * 1) + (3 * 2) = 10.
dot_product = np.dot(vector_a, vector_b)

print("Vector A + Vector B:", vector_sum)
print("2 * Vector A:", scaled_vector)
print("Dot product of Vector A and Vector B:", dot_product)
```

## How vectors support Python programming

Vectors support Python by turning data into numerical structures that can be processed by NumPy and AI libraries. Vector operations are used in feature data, embeddings, similarity calculations, weighted sums and deep learning tensor workflows.
